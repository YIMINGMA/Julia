## Exercise 12-1

Write a function called ```sumall``` that takes any number of arguments and returns their sum.

````julia
function sumall(args...)
    summation = 0
    for ele in args
        summation = summation + ele
    end
    return summation
end

println(sumall(1,2,3))
````

## Exercise 12-2

Write a function called ```mostfrequent``` that takes a string and prints the letters in decreasing order of frequency. Find text samples from several different languages and see how letter frequency varies between languages. Compare your results with the tables at [Wikipedia](https://en.wikipedia.org/wiki/Letter_frequencies).

````julia
function mostfrequent(str)
    d = Dict()
    for ele in str
        d[ele] = get(d, ele, 0) + 1
    end
    letter_counter = []
    for (key, val) in d
        push!(letter_counter, (key, val))
    end
    # Bubble sort
    for i in length(letter_counter):(-1):2
        for j in 1: (i -1)
            if letter_counter[j][2] < letter_counter[j+1][2]
                temp = letter_counter[j]
                letter_counter[j] = letter_counter[j+1]
                letter_counter[j+1] = temp
            end
        end
    end
    for (letter, counter) in letter_counter
        println(letter, " ", counter)
    end
end
````

````julia
mostfrequent("aaaabbbccd")
````

## Exercise 12-5

Here's another Car Talk Puzzler:

What is the longest English word, that remains a valid English word, as you remove its letters one at a time?

Now, letters can be removed from either end, or the middle, but you can't rearrange any of the letters. Every time you drop a letter, you wind up with another English word. If you do that, you're eventually going to wind up with one letter and that too is going to be an English wordone that's found in the dictionary. I want to know what's the longest word. What's the word, and how many letters does it have?

I'm going to give you a little modest example: Sprite. Ok? You start off with sprite, you take a letter off, one from the interior of the word, take the *r* away, and we're left with the word spite, then we take the *e* off the end, we're left with spit, we take the *s* off,we're left with pit, it, and I.

Write a program to find all words that can be reduced in this way, and then find the
longest one.

This exercise is a little more challenging than most, so here are
some suggestions:

1. You might want to write a function that takes a word and computes an array of all the words that can be formed by removing one letter. These are the "children of the word".
2. Recursively, a word is reducible if any of its children are reducible. As a base case, you can consider the empty string reducible.
3. The word list I provided, ```words.txt```, doesn't contain single letter words. So, you might want to add ```"I"```, ```"a"```, and the empty string.
4. To improve the performance of your program, you might want to memoize the words that are known to be reducible.

````julia
using DelimitedFiles

cd("C:\\Users\\76493\\Desktop\\Codes\\Think-Julia-How-to-Think-Like-A-Computer-Scientist\\Data")
words_array = dropdims(readdlm("words.txt"), dims=2);

push!(words_array, "a");
push!(words_array, "i");
push!(words_array, "");
````

````julia
function get_children(word)
    word = collect(word)
    children = []
    for i in 1: length(word)
        copy = word[:]
        push!(children, String(deleteat!(copy, i)))
    end
    return children
end
````

````julia
println(get_children("hello"))
println(get_children("a"))
````

````julia
d = Dict("" => true, "a" => true, "i" => true)
function isreducible(word)
    if word in keys(d) 
        return d[word]
    else
        children = get_children(word)
        result = false
        for child in children
            if child in words_array && isreducible(child)
                result = true
                break
            end
        end
        d[word] = result
        return result
    end
end
````

````julia
reducible = []
the_word = []
the_length = 0
for word in words_array
    if isreducible(word)
        push!(reducible, word)
        if length(word) > the_length
            the_word = [word]
            the_length = length(word)
        elseif length(word) == the_length
            push!(the_word, word)
        end
    end
end
````
